import pandas as pd

```python
df = pd.read_csv("googleplaystore.csv")
df
```

```python
df.columns
```

```python
df.head(5)
```

* 2. Display Category with highest number of apps registered. *

```python
app_df = df['Category'].value_counts()
app_df
```

```python
high_registration = app_df.idxmax()
high_registration
high_registration_count = app_df.max()
high_registration_count
print("high_registration in   :",high_registration)
print("high_registration count:",high_registration_count)
```

* 3. What is the average ratings for all the apps? *

```python
avg_rating = df['Rating'].mean()
avg_rating
```

* 4. Total number of apps which even supports Andriod version 2.3 and up. *

```python
android_version_nan = df['Android Ver'].isna()

android_version = df[~android_version_nan & df['Android Ver'].str.startswith('2.3 and up')]
android_version
```

```python
android_version = df[df['Android Ver']== '2.3 and up']
android_version
```

* 5. Impute the missing values available in Content Rating column with respective measure of central tendency. *

```python
content_rating = df['Content Rating'].isnull().value_counts()
content_rating 
```
